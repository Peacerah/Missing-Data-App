```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
# Read the complete dataset with missing data
train_10per = pd.read_csv("Apapa_Tidal_Export_10PerMiss.csv")

# init setup
from pycaret.regression import *
s = setup(train_10per, target = "Water_Level")

#model training and selection
best = compare_models()

#analyze best model
evaluate_model(best)

# extract the data with missing column 
test_10per = train_10per[train_10per['Water_Level'].isnull()]

#predict on new data
predictions = predict_model(best, data = test_10per)

# extract the data with complete case and drop the missing data
train_10per_noNull = train_10per.dropna(how='any')

# drop the column with missing data in the predicted output
predictions_1 =predictions.drop(columns="Water_Level" )

# Rename the label column in the predicted output with the Water_Level
predictions_1.rename(columns={"Label":"Water_Level"} ,inplace=True)

# Merging the complete case with the predicted to give a dataset
frame = [train_10per_noNull,predictions_1]
output_10 = pd.concat(frame)
output_10 = output_10.sort_index()
```
